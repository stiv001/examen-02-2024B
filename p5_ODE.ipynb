## Comparación con la solución real
La solución real de la ecuación diferencial es:
$y(t)=t^2 +1/3e^{-5t}$

* Grafique la solución real y la aproximación obtenida con el método de Euler en una misma gráfica.



# Errores
Calcule el eeror relativo para diferentes valores de N.

```python
import math

def f(t: float, y: float) -> float:
    return (-5) * y + 5 * (t ** 2) + 2 * t

a = 0
b = 1
y_t0 = 1 / 3
N = 10

resultado = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)  
t = resultado[1]
h = resultado[2]

y_exact = []
for ti in t:
    y_exact.append((ti ** 2) + (1 / 3) * math.exp(-5 * ti))

error = []
for i in range(len(y)):
    error.append(y[i] - y_exact[i])

print("t:", t)
print("y:", y)

error_relativo = []
for i in range(len(error)):
    error_relativo.append(abs(error[i] / y_exact[i]))

print("Error relativo:", error_relativo)

promedio_error_relativo = sum(error_relativo) / len(error_relativo)
print("Promedio del error relativo:", promedio_error_relativo)
print("Promedio del error relativo: {:.2e}".format(promedio_error_relativo))
```

```python
import matplotlib.pyplot as plt

y_aprox, t, h = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)

y_real = [(ti ** 2) + (1/3) * math.exp(-5 * ti) for ti in t]

plt.figure(figsize=(8, 5))
plt.plot(t, y_real, label="Y_real ", linestyle="dashed", color="r")
plt.plot(t, y_aprox, label="Y_aproximado (Euler)", marker="o", color="g")
plt.xlabel("t")
plt.ylabel("Y(t)")
plt.title("Comparación entre Y_real y Y_aprox")
plt.legend()
plt.grid(True)
plt.show()
```
